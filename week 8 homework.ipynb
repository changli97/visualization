```python
@ipywidgets.interact(btype=gdf_apts['Fire_Alarms'].unique())
def my_geopandas_plot(btype):
    gdf_apts_subset = gdf_apts.loc[gdf_apts['Fire_Alarms']==btype] # our subsetted dataframe

    fig, ax = plt.subplots(1,1,figsize=(10,10))

    ax.set_xlim(-88.22, -88.26)
    ax.set_ylim(40.10, 40.16)

    gdf_councils.plot(ax=ax, alpha=0.5, edgecolor='k')
    gdf_apts_subset.plot(column='Fire_Alarms', ax=ax, cmap='rainbow', categorical=True, legend=True) # look what changes

    ctx.add_basemap(ax=ax, crs=gdf_councils.crs.to_string())

    plt.show()
```
